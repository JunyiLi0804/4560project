```python
import numpy as np

resnet18_labels_test = np.load('resnet18_labels_test.npy')
resnet18_probs_test = np.load('resnet18_probs_test.npy')
resnet18_predictions = resnet18_probs_test.argmax(axis=1)

resnet50_labels_test = np.load('resnet50_labels_test.npy')
resnet50_probs_test = np.load('resnet50_probs_test.npy')
resnet50_predictions = resnet50_probs_test.argmax(axis=1)


resnet50pre_labels_test = np.load('resnet50pre_labels_test.npy')
resnet50pre_probs_test = np.load('resnet50pre_probs_test.npy')
resnet50pre_predictions = resnet50pre_probs_test.argmax(axis=1)
```

```python
def topk_accuracy(probs, labels, k=3):
  n = len(probs)
  cnt = 0
  for i in range(n):
    topk = np.argsort(-probs[i])[:k]
    if labels[i] in topk:
      cnt += 1
  return cnt/n
```

```python
print('resnet18 top1: ', topk_accuracy(resnet18_probs_test, resnet18_labels_test, 1))
print('resnet18 top3: ', topk_accuracy(resnet18_probs_test, resnet18_labels_test, 3))
print('resnet18 top5: ', topk_accuracy(resnet18_probs_test, resnet18_labels_test, 5))

print('resnet50 top1: ', topk_accuracy(resnet50_probs_test, resnet50_labels_test, 1))
print('resnet50 top3: ', topk_accuracy(resnet50_probs_test, resnet50_labels_test, 3))
print('resnet50 top5: ', topk_accuracy(resnet50_probs_test, resnet50_labels_test, 5))


print('resnet50pre top1: ', topk_accuracy(resnet50pre_probs_test, resnet50_labels_test, 1))
print('resnet50pre top3: ', topk_accuracy(resnet50pre_probs_test, resnet50_labels_test, 3))
print('resnet50pre top5: ', topk_accuracy(resnet50pre_probs_test, resnet50_labels_test, 5))
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score

avg_mode = ['micro', 'macro']

for mode in avg_mode:
  print('resnet18 precision({}):'.format(mode), precision_score(resnet18_predictions, resnet18_labels_test, average=mode))
  print('resnet18 recall({}):'.format(mode), recall_score(resnet18_predictions, resnet18_labels_test, average=mode))
  print('resnet18 f1({}):'.format(mode), f1_score(resnet18_predictions, resnet18_labels_test, average=mode))

  print('resnet50 precision({}):'.format(mode), precision_score(resnet50_predictions, resnet50_labels_test, average=mode))
  print('resnet50 recall({}):'.format(mode), recall_score(resnet50_predictions, resnet50_labels_test, average=mode))
  print('resnet50 f1({}):'.format(mode), f1_score(resnet50_predictions, resnet50_labels_test, average=mode))
  
  print('resnet50pre precision({}):'.format(mode), precision_score(resnet50pre_predictions, resnet50pre_labels_test, average=mode))
  print('resnet50pre recall({}):'.format(mode), recall_score(resnet50pre_predictions, resnet50pre_labels_test, average=mode))
  print('resnet50pre f1({}):'.format(mode), f1_score(resnet50pre_predictions, resnet50pre_labels_test, average=mode))
  
  print('')
```
